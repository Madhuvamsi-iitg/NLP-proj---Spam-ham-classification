```python
import pandas as pd
import numpy as np
df = pd.read_csv('/content/spam.csv',encoding='latin-1')
```

```python
df = df.rename(columns={'v1':'label','v2':'text'})
```

```python
df = df.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1)
```

```python
df = df.dropna()
```

```python
df
```

```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
nltk.download('wordnet')
```

```python
corpus = []
for i in range (0,len(df)):

  review = re.sub('[^a-zA-Z]',' ',df['text'][i]) #it says that apart from a to z and A to Z,if you find any special character,replace with blank.
  review = review.lower()
  review = review.split()
  review = [lemmatizer.lemmatize(word) for word in review ]
  review = ' '.join(review)
  corpus.append(review)
```

```python
[[i,j,k] for i,j,k in zip(list(map(len,corpus)),corpus,df['text']) if i<1]
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess #it converts a document into a list of lowercase tokens,ignoring that are too short or too long.
```

```python
import nltk
nltk.download('punkt')
```

```python
words=[]
for sent in corpus:
  sent_token = sent_tokenize(sent)
  for sent in sent_token:
    words.append(simple_preprocess(sent)) #here for every sentence, i am getting list of words
```

here if we apply word2vec, we will get 100 dimensional vectors for each and every word in the sentence. but we want for every sentence,a 100 dimensional vector.

```python
#here we use avgword2vec
def avg_word2vec(doc):
  return np.mean([model.wv[word] for word in doc if word in model.wv.index_to_key],axis=0)
```

```python
!pip install tqdm
```

```python
from tqdm import tqdm
```

```python
#apply for entire sentences
import numpy as np
x=[]
for i in tqdm(range(len(words))):
  x.append(avg_word2vec(words[i]))
```

```python
x
```

```python
len(x)
```

```python
#independent features
x_new = np.array(x)
```

```python
# Inspect the shape of each element
for i, arr in enumerate(x):
    print(f"Array {i} shape: {arr.shape}")

# Define the target shape
target_shape = (100,)  # Assuming you want each array to have 100 elements

# Fix the shapes of the arrays
x_fixed_shape = [np.resize(arr, target_shape) if arr.shape != target_shape else arr for arr in x]

# Verify the shapes after fixing
for i, arr in enumerate(x_fixed_shape):
    print(f"Fixed Array {i} shape: {arr.shape}")

# Convert the list of fixed arrays to a NumPy array
x_new = np.array(x_fixed_shape)
print(f"Combined array shape: {x_new.shape}")
```
