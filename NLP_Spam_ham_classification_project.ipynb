```python
import pandas as pd
df = pd.read_csv('/content/spam.csv',encoding = 'latin-1')
```

```python
df = df.rename(columns={'v1':'label','v2':'text'})
```

```python
df = df.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1)
```

```python
df = df.dropna()
```

```python
df.isna().sum()
```

```python
df
```

```python
#covert label column to numerical variables
y = df['label'].map({'ham':1,'spam':0})
```

```python
y
```

```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
wordlemmatize=WordNetLemmatizer()
```

```python
corpus = []
for i in range (0,len(df)):

  review = re.sub('[^a-zA-Z]',' ',df['text'][i]) #it says that apart from a to z and A to Z,if you find any special character,replace with blank.
  review = review.lower()
  review = review.split()
  review = [wordlemmatize.lemmatize(word) for word in review if not word in stopwords.words('english')]
  review = ' '.join(review)
  corpus.append(review)
```

```python
corpus
```

**create Bag of words**

```python
from sklearn.feature_extraction.text import CountVectorizer
#for binary BOW enable binary=True
cv = CountVectorizer(max_features=2500,binary=True,ngram_range=(1,2))
X = cv.fit_transform(corpus).toarray() #independent features
```

```python
import numpy as np
np.set_printoptions(edgeitems=30,linewidth=100000,
                    formatter=dict(float=lambda x: "%.3g" % x))
X
```

```python
cv.vocabulary_
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_detect_model = MultinomialNB().fit(X_train,y_train)
```

```python
y_pred = spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
accuracy = accuracy_score(y_test,y_pred)
confusion_m = confusion_matrix(y_test,y_pred)
```

```python
print(accuracy)
```

```python
print(confusion_m)
```

```python
print(classification_report(y_test,y_pred))
```

**create TF-IDF model**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tv =TfidfVectorizer(max_features=2500,ngram_range=(1,2))
x = tv.fit_transform(corpus).toarray()
```

```python
x
```

```python
tv.vocabulary_
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_tfidf_model = MultinomialNB().fit(X_train,y_train)
```

```python
#prediction
y_pred = spam_tfidf_model.predict(X_test)
```

```python
score = accuracy_score(y_test,y_pred)
cm = confusion_matrix(y_test,y_pred)
```
